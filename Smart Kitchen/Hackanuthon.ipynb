```python
from ultralytics import YOLO  # Import YOLO class

# Load the YOLOv8 model
model = YOLO('/content/drive/MyDrive/extracted_folder/best_model.pt')  # 'n' means nano version (smallest), change to 's', 'm', 'l' as needed
```

```python

# Train the model
model.train(data="/content/drive/MyDrive/extracted_folder/data.yaml", epochs= 10, imgsz=640, batch=8)
model_path = "/content/drive/MyDrive/extracted_folder/best_model.pt"
model.save(model_path)
print(f"✅ Model saved at: {model_path}")
```
